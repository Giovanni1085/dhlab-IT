`MASTER=local[30] IPYTHON=1 IPYTHON_OPTS="notebook --profile nbserver" ./spark-1.5.1-bin-hadoop2.6/bin/pyspark --driver-memory 20g`

Let's go through the parameters quickly :
- `MASTER=local[30]` : just a local spark system, not distributed, but with 30 threads
- `IPYTHON=1` : use the IPython system
- `IPYTHON_OPTS="notebook --profile nbserver"` : use the notebook system of IPython with the local profile `nbserver` (that needs to be created before see `Introduction`) allowing foreign connections
- `--driver-memory 20g` : usually `--executor-memory` would be used to set the amount of memory per executor. But here, since we are in local mode everything is handled by the driver, which needs enough memory hence the parameter

# Loading data

## Loading the files

Since our data is a bunch of `.xml` files that need to be parsed, we are loading them separately to a `RDD` with `sc.wholeTextFiles`.

Note however, that by default the RDD will be separated in only two different partitions which is greatly inefficient. Hence, we force the use of the total number of threads with the `minPartitions` parameter.

NB : the `use_unicode=False` bit is needed for the XML parser to work properly apparently (not too sure about that)

```python
data = sc.wholeTextFiles('~/dh-shared/GDL/',use_unicode=False)
data.getNumPartitions()
```

```python
data = sc.wholeTextFiles('~/dh-shared/GDL/',use_unicode=False,minPartitions=sc.defaultParallelism)
data.getNumPartitions()
```

We can verify the number of files we analyze. In our data, it's roughly one file per year, showing the span of the 160 years of newspapers

```python
data.count()
```

## Extracting the articles

Now, we need to parse our xml files. First we write a helper function that process a single `<article>` xml node.

```python
import xml.etree.ElementTree as ET

def get_article_text(a):
	article_text = ""
	txt_parts = map(lambda x : x.find('full_text').text, a.findall('entity'))
	txt_parts = filter(lambda x : x, txt_parts) #remove None element
	return " ".join(txt_parts)
```

Here is the main tricky part. Our variable `data` is a RDD of elements `(str, str)` where the first element is the title of the file and the second its content.

- `.map(lambda x : ET.fromstring(x[1]))` : select only the content part
- `.flatMap(lambda xml : xml.findall('article') )` : output the list of the separate articles and concatenate them
- `.map(get_article_text)` : extract the article text with the helper function
- `.repartition(sc.defaultParallelism)` : spread the data evenly between the different partitions for more efficient proessing later
- `articles.cache()` : save this RDD in memory so we do not have to recompute it several times

```python
articles = data.map(lambda x : ET.fromstring(x[1])) \
			.flatMap(lambda xml : xml.findall('article') ) \
			.map(get_article_text) \
			.repartition(sc.defaultParallelism)
articles.cache()
```

Remember that Spark is based on lazy operations. Until we actually ask for a result nothing is done. So this simple count command will actually launch all the loading and parsing of the data. Since we asked our RDD to be cached, subsequent calls will be way faster because the processed data stays in memory.

```python
articles.count()
```

# Processing the data

Now that our data is ready we can just apply operations on it, for instance the ever-classic example of counting words. Documentation on these lines can be found everywhere on any map-reduce tutorial.

```python
from operator import add

counts = articles.flatMap(lambda x: x.split(' ')) \
			.map(lambda x: (x, 1)) \
			.reduceByKey(add) \
			.sortBy(lambda x: -x[1]) \
			.collect()
```

This took roughly 50s to count all the words of the GDL. And then 30s to regroup them in the collect operation to a local python list.

Looks like we have **10 Millions** elements and we can output the first 10 elements for instance

```python
print len(counts)
print counts[:10]
```

Let us save our data in a file if we need it later

```python
import cPickle as pickle
pickle.dump( counts, open( "1-grams.p", "wb" ) )
```

## BONUS, let us try for the n-grams.

```python
def ngrams(input, n):
    splitted = input.split(' ')
    splitted = filter(lambda x : len(x)>0, splitted)
    output = list()
    for i in range(len(splitted)-n+1):
        output.append(' '.join(splitted[i:i+n]))
    return output
```

```python
articles.flatMap(lambda x: ngrams(x, 1) ) \
    .map(lambda x: (x, 1)) \
    .reduceByKey(add) \
    .sortBy(lambda x: -x[1]) \
    .saveAsPickleFile('1-grams')
articles.flatMap(lambda x: ngrams(x, 2) ) \
    .map(lambda x: (x, 1)) \
    .reduceByKey(add) \
    .sortBy(lambda x: -x[1]) \
    .saveAsPickleFile('2-grams')
```

Here we generate the grams and output them directly from the RDD to the filesystem. That way, each executor save its own data on a separate file. Can be better when the data gets quite big and one giant python list is not going to be able to manage it anymore.

The generation of the 1-grams took 1.2min and saving them 30s

For the 2-grams :
- the reducing operation 3.5 min
- the sorting of the elements took 3.8 min
- the saving part took 4.7min (!). This is quite high, and if we look at the saving, most of the work was done by just one executor so our data got imbalanced in the process. Indeed 75% of the 2Go output were written by the same executor.

Still, despite the lenghty saving process, **generating the 2-grams for the GDL corpus took roughly 10 minutes**.

# Basic analysis

We only want the number of occurences per word, so we conccurently load our data organised as `(str,int)` tuples, and we get rid of the string key to just keep the number of occurences in order to make it manageable.

```python
numbers_occurences_1 = sc.pickleFile('1-grams',minPartitions=sc.defaultParallelism).map(lambda x: x[1]).collect()
numbers_occurences_2 = sc.pickleFile('2-grams',minPartitions=sc.defaultParallelism).map(lambda x: x[1]).collect()
```

```python
print "Individual 1-grams : "+str(len(numbers_occurences_1))
print "Individual 2-grams : "+str(len(numbers_occurences_2))
print "Number of words : "+str(reduce(add,numbers_occurences_1))
print "Number of 2-grams : "+str(reduce(add,numbers_occurences_2))
```

## 1-grams

```python
plt.hist(map(lambda x:x[1], counts), bins=40, log=True)
plt.title('Words repartition')
plt.ylabel('Number of words')
plt.xlabel('Number of occurences')
```

A small histogram even in log-scale show the extremely skewed distribution of the data. So we want to investigate further in this direction. We need a pie chart of the proportion of words represented by their occurences and what are their weights in the global representation of the words.

```python
bins = [0,1,5,20,100,1000,10000,100000,1000000,Inf]
words_per_bins = list()
occurences_per_bins = list()
labels = list()
for i in range(len(bins)-1):
    elements_in_bin = filter(lambda x: x>bins[i] and x<=bins[i+1],numbers_occurences_1)
    words_per_bins.append(len(elements_in_bin))
    occurences_per_bins.append(reduce(add,elements_in_bin))
    labels.append("{} < _ <= {}".format(bins[i],bins[i+1]))
```

```python
plt.pie(words_per_bins, shadow=True, labels=labels, autopct='%1.1f%%')
plt.title("Number of words by their number of occurences")
```

```python
plt.pie(occurences_per_bins, shadow=True, labels=labels, autopct='%1.1f%%')
plt.title("Weight of each group in the total number of words")
```
